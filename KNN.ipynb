## Introduction

In this project, we explore two machine learning methods: K-Nearest Neighbors and K-Means clustering.

We apply these methods to the MNIST dataset, which contains handwritten digit images from 0 to 9.

The goal of this project is to:

- use KNN to classify digits and study how the value of K affects accuracy  
- use K-Means to group similar images into clusters  
- explore a variation of KNN called weighted KNN

By comparing these methods, we can better understand how different algorithms work on image data.

```python
import pandas as pd
import numpy as np

# read the csv file
data = pd.read_csv("mnist_train.csv")

# convert to numpy array
Data = np.array(data)

# check shape
Data.shape
```

```python
# separate labels and pixels

y = Data[:,0]        # labels
X = Data[:,1:]       # image pixels

print(X.shape)
```

Now we split the data into training and testing sets.

We will use 10,000 images for testing.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=10000,
    random_state=42,
    stratify=y
)
```

## KNN Classification

We now apply the K-Nearest Neighbors algorithm.

The idea is:

For each test point:
1. find the K nearest training points
2. take the most common label

```python
from sklearn.neighbors import KNeighborsClassifier

# try K = 3
MyKNN = KNeighborsClassifier(n_neighbors=3)

MyKNN.fit(X_train, y_train)
```

```python
# predict on test data
pred = MyKNN.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred)
```

Now we test different values of K.

```python
k_values = [1,3,5,7,9]
acc_list = []

for k in k_values:
    
    MyKNN = KNeighborsClassifier(n_neighbors=k)
    MyKNN.fit(X_train, y_train)
    
    pred = MyKNN.predict(X_test)
    
    acc = accuracy_score(y_test, pred)
    acc_list.append(acc)
    
    print("K =", k, "accuracy =", acc)
```

```python
import matplotlib.pyplot as plt

plt.plot(k_values, acc_list, 'o-')
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.title("KNN Accuracy")
plt.show()
```

## K-Means Clustering

Now we apply K-Means clustering.

We use K = 10 clusters because there are 10 digits.

```python
from sklearn.cluster import KMeans

MyKM = KMeans(n_clusters=10, random_state=42)

MyKM.fit(X_train)
```

```python
# look at centroids
centroids = MyKM.cluster_centers_

centroids.shape
```

```python
# plot centroids

plt.figure(figsize=(8,4))

for i in range(10):
    plt.subplot(2,5,i+1)
    
    img = centroids[i].reshape(28,28)
    
    plt.imshow(img, cmap='gray')
    plt.axis('off')

plt.show()
```

We can see that many centroids look like digits.

This means K-Means is capturing the structure of the data.

## Weighted KNN

Now we try weighted KNN.

In this method, closer neighbors have more influence.

Weight formula:

$$
w = \frac{1}{d}
$$

```python
# normal KNN
knn1 = KNeighborsClassifier(n_neighbors=3)
knn1.fit(X_train, y_train)

pred1 = knn1.predict(X_test)

acc1 = accuracy_score(y_test, pred1)
print("Normal KNN:", acc1)
```

```python
# weighted KNN
knn2 = KNeighborsClassifier(n_neighbors=3, weights='distance')
knn2.fit(X_train, y_train)

pred2 = knn2.predict(X_test)

acc2 = accuracy_score(y_test, pred2)
print("Weighted KNN:", acc2)
```

We see that weighted KNN gives slightly better accuracy.

This is because closer neighbors are more important.

## Conclusion

In this project, we applied both KNN and K-Means algorithms to the MNIST dataset. We found that KNN gives very high accuracy when classifying handwritten digits.

K-Means clustering helped us understand the structure of the data by grouping similar images together. The cluster centroids often look like real digits, which shows that the algorithm captures important patterns.

Finally, we tested weighted KNN and found that it gives slightly better accuracy than standard KNN. This shows that closer neighbors provide more useful information for prediction.
